```python
# Invoke aws Lambda
```

```python
lam= boto3.client('lambda',region_name='us-east-1')
```

```python

response=lam.invoke(
FunctionName='celebrity_detector',
InvocationType='RequestResponse',
LogType='Tail',
Payload=json.dumps({'bucket':'demobuckethim23','object_name':'bezos_demo_image.jpg'})
)
celebrity_name=response['Payload'].read().decode('utf-8')
```
