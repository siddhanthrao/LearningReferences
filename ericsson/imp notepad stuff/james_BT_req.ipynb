# Handling functions for 5 data's

# Data 1 same date open & close

```python
def func1(data):
    
    data['inc_assignment_start_date Timestamp']=data['inc_assignment_start_date'].dt.time
    data['inc_assignment_start_date Timestamp2']=data['inc_assignment_start_date Timestamp'].apply(str)
    data['inc_assignment_start_date Timestamp3']=data['inc_assignment_start_date Timestamp2'].apply(lambda x: int(x[0:2])*60 + int(x[3:5]))
    data['inc_assignment_end_date Timestamp']=data['inc_assignment_end_date'].dt.time
    data['inc_assignment_end_date Timestamp2']=data['inc_assignment_end_date Timestamp'].apply(str)
    data['inc_assignment_end_date Timestamp3']=data['inc_assignment_end_date Timestamp2'].apply(lambda x: int(x[0:2])*60 + int(x[3:5]))

    
    # same day open & closure but not a business day it is
    data_1=data[data['BHS']==0]

    # reamining data
    data_2=data[data['BHS']!=0]
    
    data_2['BHS']=data_2['BHS']-9
    
    data_2['BHS']=np.where(data_2['inc_assignment_start_date Timestamp3']> 1020, 0,
                       np.where(((data_2['inc_assignment_start_date Timestamp3']<= 1020) & (data_2['inc_assignment_start_date Timestamp3'] >= 480) & ((data_2['inc_assignment_end_date Timestamp3'] <= 1020))),(data_2['inc_assignment_end_date Timestamp3']-data_2['inc_assignment_start_date Timestamp3'])/60,
                                np.where(((data_2['inc_assignment_start_date Timestamp3']<= 1020) & (data_2['inc_assignment_start_date Timestamp3'] >= 480) & ((data_2['inc_assignment_end_date Timestamp3'] > 1020))),(1020-data_2['inc_assignment_start_date Timestamp3'])/60,
                                        np.where(((data_2['inc_assignment_start_date Timestamp3']<480) & (data_2['inc_assignment_end_date Timestamp3']<480)),0,
                                                np.where(((data_2['inc_assignment_start_date Timestamp3']<480) & (data_2['inc_assignment_end_date Timestamp3']>=480) & (data_2['inc_assignment_end_date Timestamp3']<=1020)), (data_2['inc_assignment_end_date Timestamp3']-480)/60,
                                                         np.where(((data_2['inc_assignment_start_date Timestamp3']<480) & (data_2['inc_assignment_end_date Timestamp3']> 1020)),9,'Not handled'))))))
                                                 
                                                         

    data_new=data_1.append([data_2])
    
    data_new['BHS']=data_new['BHS'].astype(float)
    
    return data_new
```

# Data 3 Different day open & close + open (business day) + closed (non business day)


```python
def func3(data):
    
    data['BHS']=data['BHS']-9
    data['inc_assignment_start_date Timestamp']=data['inc_assignment_start_date'].dt.time
    data['inc_assignment_start_date Timestamp2']=data['inc_assignment_start_date Timestamp'].apply(str)
    data['inc_assignment_start_date Timestamp3']=data['inc_assignment_start_date Timestamp2'].apply(lambda x: int(x[0:2])*60 + int(x[3:5]))

    
    data['BHS']=np.where(data['inc_assignment_start_date Timestamp3']> 1020, data['BHS']+0,
                   np.where(((data['inc_assignment_start_date Timestamp3']<= 1020) & (data['inc_assignment_start_date Timestamp3'] >= 480)),data['BHS']+(1020-data['inc_assignment_start_date Timestamp3'])/60,
                            np.where(data['inc_assignment_start_date Timestamp3']< 480,data['BHS']+9,'Not handled yet')))
    
    data['BHS']=data['BHS'].astype(float)
    
    return data
```
